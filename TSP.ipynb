```python
import numpy as np

def reduce_matrix(distance_matrix):
    reduced_matrix = distance_matrix.copy()
    for row in reduced_matrix:
        row_min = np.min(row)
        row -= row_min

    for col in reduced_matrix.T:
        col_min = np.min(col)
        col -= col_min

    return reduced_matrix

def calculate_lower_bound(reduced_matrix):
    row_min = np.min(reduced_matrix, axis=1)
    col_min = np.min(reduced_matrix, axis=0)
    lower_bound = np.sum(row_min) + np.sum(col_min)
    return lower_bound

def tsp_branch_and_bound(distance_matrix):
    reduced_matrix = reduce_matrix(distance_matrix)
    current_node = 0
    current_path = []
    current_cost = 0
    lower_bound = calculate_lower_bound(reduced_matrix)
    best_solution = None
    best_cost = float('inf')

    def expand_branch():
        nonlocal current_node, current_path, current_cost
        for i in range(len(distance_matrix)):
            if i not in current_path and reduced_matrix[current_node, i] != 0:
                current_path.append(i)
                current_cost += distance_matrix[current_node, i]
                current_node = i
                break

    def update_cost():
        nonlocal current_cost
        current_cost = np.sum(distance_matrix[current_path[:-1], current_path[1:]])

    def update_lower_bound():
        nonlocal lower_bound
        row_min = np.min(reduced_matrix[current_node, :], axis=0)
        col_min = np.min(reduced_matrix[:, current_node], axis=0)
        lower_bound = current_cost + np.sum(row_min) + np.sum(col_min)

    while current_node != 0 or len(current_path) < len(distance_matrix):
        if lower_bound > best_cost:
            continue

        if len(current_path) == len(distance_matrix):
            if current_cost < best_cost:
                best_solution = current_path.copy()
                best_cost = current_cost
                return

        expand_branch()
        update_cost()
        update_lower_bound()

    return best_solution, best_cost

if __name__ == '__main__':
    # Example distance matrix
    distance_matrix = np.array([[0, 10, 15, 20],
                                [10, 0, 35, 25],
                                [15, 35, 0, 30],
                                [20, 25, 30, 0]])

    best_solution, best_cost = tsp_branch_and_bound(distance_matrix)
    print(f"Optimal solution: {best_solution}")
    print(f"Optimal cost: {best_cost}")
```

```python
marked = set()
n = int(input())
c = []
cmin = float('inf')

for _ in range(n):
    temp = list(map(int, input().split()))
    cmin = min(cmin, sorted(temp)[1])
    c.append(temp)

dp = {}  # Memoization table to store solutions

def Try(k, f, f_min, marked, x):
    if f + c[x[n - 1]][x[0]] >= f_min:  # Early termination check
        return f_min

    key = (k, f)  # Key for memoization
    if key in dp:  # Check if solution already computed
        return dp[key]

    for v in range(n):
        if v not in marked:
            x[k] = v
            f += c[x[k - 1]][x[k]]
            marked.add(v)

            if k == n - 1:
                f_min = solution(f, f_min, x)
            else:
                g = f + cmin * (n - k + 1)
                if g < f_min:
                    f_min = Try(k + 1, f, f_min, marked, x)

            marked.remove(v)
            f -= c[x[k - 1]][x[k]]

    dp[key] = f_min  # Store computed solution in memoization table
    return f_min

def solution(f, f_min, x):
    if f + c[x[n - 1]][x[0]] < f_min:
        f_min = f + c[x[n - 1]][x[0]]
    return f_min

if __name__ == '__main__':
    x = [0] * n
    marked.add(0)
    result = Try(1, 0, float('inf'), marked, x)
    print(result)
```
